```python
from transformers import pipeline

mentor = pipeline(
    "text2text-generation",
    model="google/flan-t5-large"
)
```

```python
def domain_mentor_bot(question, domain="General Software Engineering"):
    prompt = f"""
You are a senior mentor in {domain}.

Explain the following clearly.
Include:
1. Simple explanation
2. Example
3. 2 short practice questions

Question:
{question}
"""

    result = mentor(prompt, max_length=350)
    return result[0]["generated_text"]
```

```python
mentor = pipeline(
    "text2text-generation",
    model="google/flan-t5-base"
)
```

```python
def domain_mentor_bot(question, domain="SQL"):
    prompt = f"""
You are a senior software mentor.

Task:
Explain the concept briefly and give ONE correct example.

Rules:
- Do NOT repeat words
- Do NOT generate multiple SQL queries
- Use plain English
- One SQL query only

Domain: {domain}
Question: {question}

Answer format:
Explanation:
Example:
Practice Questions:
"""

    result = mentor(
        prompt,
        max_new_tokens=180,
        do_sample=False,
        num_beams=1,
        repetition_penalty=1.5
    )

    return result[0]["generated_text"]
```

```python
print(domain_mentor_bot("Explain SQL INNER JOIN", "SQL"))
```

```python
print(domain_mentor_bot("Explain SQL INNER JOIN", "SQL"))
```

```python
import pandas as pd
from datetime import datetime

logs = []

def log_usage(question, domain):
    logs.append({
        "question": question,
        "domain": domain,
        "timestamp": datetime.now()
    })

# Example usage
log_usage("Explain SQL JOIN", "SQL")
log_usage("Explain recursion", "Python")

df_logs = pd.DataFrame(logs)
df_logs
```

```python
import matplotlib.pyplot as plt

df_logs['domain'].value_counts().plot(kind='bar')
plt.title("Questions Asked per Domain")
plt.xlabel("Domain")
plt.ylabel("Count")
plt.show()
```
